Consider the following ODE: $\frac{dy}{dt} = e^y - sin(y)$. How can you write a function to approximate the solution? Use Euler's Method, where $y(t + \Delta{t}) = y(t) + \Delta{t}f(y,t)$.

(In DiffEQ last year, we used Picard's Method of Successive Approximations to find $y_1, y_2, ..., y_n$ when given an ODE and an initial condition, which uses the formula: <br>
$y_n(x) = y_0 + \int\limits_{x_0}^{x}f(t, y_{n-1}(t)) dt$ <br>
This is what I was thinking of when I said we had used something similar to approximate points of a solution to an ODE.)

**Now, going back to Euler's Method... $y(t + \Delta{t}) = y(t) + f(y,t)\Delta{t}$.**

Consider an ODE: $\frac{dy}{dt} = t$.

```python
#define ODE
def F1(t, y):
    return t
```

```python
#define Euler's method
#diffeq solver THAT WORKS

def eulers_method(t0, y0, h, tN):
    
    y_soln = [y0]
    
    time_values = np.arange(t0, tN, h)
    
    #print(np.arange(t0, tN, h))
    
    y_old = y0
    
    for t in time_values:
        y_new = y_old + (F1(t, y_old))*h
        y_soln += [y_new]
        y_old = y_new
        
    return y_soln, time_values
```

```python
y_soln, time_values = eulers_method(-4, 0, .1, 4);
```

```python
plt.plot(time_values, y_soln[:-1], 'k')

plt.title("Approximate Solution Curve")
```

```python
#graph of exact solution to previous ODE
t = np.linspace(-4, 4, 100)

y = (1/2)*t**2

plt.plot(t, y, 'k', label = "Exact Solution Curve")
plt.title("Exact Solution Curve")
plt.show()
```

```python
# plot the matrix.
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
im = ax.imshow(matrix_5x7)
plt.show()
```
