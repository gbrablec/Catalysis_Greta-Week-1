#### 1) Define a function that takes in a vector and returns a vector.

*For simplicity, we will be working in 3-space.*

```python
def vector_return_vector():
    """ This function adds two vectors together. """
    
    #user generates vector_user
    print("Please enter an integer value between -10 and 10.")
    
    a = int(input("Enter a value for x component: "))
    while a < -10 or a > 10:
            print("Entered value not allowed. Please enter an integer value between -10 and 10.")
            a = int(input("Enter a value for x component: "))
        
    b = int(input("Enter a value for y component: "))
    while b < -10 or b > 10:
            print("Entered value not allowed. Please enter an integer value between -10 and 10.")
            b = int(input("Enter a value for y component: "))
        
    c = int(input("Enter a value for z component: "))
    while c < -10 or c > 10:
            print("Entered value not allowed. Please enter an integer value between -10 and 10.")
            c = int(input("Enter a value for z component: "))
    
    vector_user = [a,b,c]
    print("Vector A =")
    print(vector_user)
    
    #computer generates vector_program
    d = int(random.randint(-10,10))
    e = int(random.randint(-10,10))
    f = int(random.randint(-10,10))
    vector_program = [d,e,f]
    print("Vector B =")
    print(vector_program)
    
    #compute solution vector (vector_user + vector_program)
    g = vector_user[0] + vector_program[0]
    h = vector_user[1] + vector_program[1]
    i = vector_user[2] + vector_program[2]
    vector_soln = [g,h,i]
    
    #print solution vector
    print("A + B = ")
    print(vector_soln)
```

```python
vector_return_vector();
```

#### 2) Define a function that takes in a vector and returns a scalar.

*Again... 3-space!*

```python
def vector_return_scalar():
    """ This function computes the dot product between two vectors. """
    
    #user generates vector_user
    print("Please enter an integer value between -10 and 10.")
    
    a = int(input("Enter a value for x component: "))
    while a < -10 or a > 10:
            print("Entered value not allowed. Please enter an integer value between -10 and 10.")
            a = int(input("Enter a value for x component: "))
            
        
    b = int(input("Enter a value for y component: "))
    while b < -10 or b > 10:
            print("Entered value not allowed. Please enter an integer value between -10 and 10.")
            b = int(input("Enter a value for y component: "))
            
            
    c = int(input("Enter a value for z component: "))
    while c < -10 or c > 10:
            print("Entered value not allowed. Please enter an integer value between -10 and 10.")
            c = int(input("Enter a value for z component: "))
            
    
    vector_user = [a,b,c]
    print("Vector A =")
    print(vector_user)
    
    #computer generates vector_program
    d = int(random.randint(-10,10))
    e = int(random.randint(-10,10))
    f = int(random.randint(-10,10))
    vector_program = [d,e,f]
    print("Vector B =")
    print(vector_program)
    
    #compute dot product
    dot_product = (vector_user[0] * vector_program[0]) 
    + (vector_user[1] * vector_program[1]) 
    + (vector_user[2] * vector_program[2])
    
    #print solution
    print("A dot B = ")
    print(dot_product)
```

```python
vector_return_scalar();
```

#### 3) Plot two regular (scalar) functions on the same plot.

```python
x = np.linspace(0, 5, num = 100)
F1 = ((x-4)**2) + 2
F2 = ((x-2)**2) - 4
```

```python
plt.plot(x, F1, 'r', label = "F1")
plt.plot(x, F2, 'b', label = "F2")

plt.xlabel("X")
plt.ylabel("Y")
plt.legend(loc="upper right")
plt.show()
```

#### 4) Plot a function that takes in two inputs and returns one output over a 2-D field.

```python
#hyperbolic paraboloid
def f(x,y):
    return -x**2 + y**2

x = np.linspace(-5, 5, num = 50)
y = np.linspace(-5, 5, num = 50)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
from matplotlib.pyplot import figure

#try to center figure
fig = plt.figure()
figure(figsize=(6,4), dpi=80)

ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap = 'jet')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z');
```

*imshow: displays data as an image.*

```python
import matplotlib.colorbar
plt.figure(figsize=(5,3), dpi=80)

plt.imshow(Z, cmap = 'jet')
plt.colorbar()

print("imshow: ")
plt.show() 
```

*contour: shows a three-dimensional surface on a two-dimensional plane.*

```python
import matplotlib.colorbar
plt.figure(figsize=(5,3), dpi=80)

plt.contourf(X, Y, Z, cmap = 'jet') 

plt.colorbar()

print("contour: ")
plt.show() 
```

*meshgrid: creates a rectangular grid out of an array of x values and an array of y values.*

```python
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)

X, Y = np.meshgrid(x, y)

print("meshgrid: ")
plt.plot(X, Y, 'k.')

plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
